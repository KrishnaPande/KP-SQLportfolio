```python
# âœ… Option 1: Loop over all columns and print unique values

for col in df.columns:
    print(f"Column: {col}")
    print(df[col].unique())
    print(len(df[col].unique()))
    print("-" * 30)


# âœ… Option 2: Save unique values in a dictionary

unique_values = {}

for col in df.columns:
    unique_values[col] = df[col].unique()

# To see the unique values for 'Contract'
print(unique_values['Contract'])
# This way, you can store all unique values and access them later.
```

ðŸ” What this does:

1. Loops over each column in your DataFrame
2. Prints the column name
3. Prints the unique values in that column
4. Adds a separator line for readability

```python
yes_no_cols = ['partner', 'dependents', 'phoneservice', 'paperlessbilling', 'churn']

for col in yes_no_cols:
    df[col] = df[col].map({'Yes': True, 'No': False})
```

| Problem                                                             | Solution                                             |
| ------------------------------------------------------------------- | ---------------------------------------------------- |
| Inconsistent casing (`'CustomerID'`, `'gender'`, `'SeniorCitizen'`) | Make all lowercase for consistency                   |
| Spaces in column names (`'Total Charges'`)                          | Replace with `_` for easy access                     |
| Special characters                                                  | Remove or replace to avoid syntax errors             |
| Readability                                                         | You want names that are clean, short, and consistent |


```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
cat_cols = ['gender', 'multiplelines', 'internetservice', 'onlinesecurity', 'onlinebackup',
            'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies',
            'contract', 'paymentmethod']

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
df.dtypes
```

```python
df['seniorcitizen'].unique()  # Should be [0, 1]
df['seniorcitizen'] = df['seniorcitizen'].astype(bool)
```

```python
df.dtypes
```

```python
binary_cols = ['partner', 'dependents', 'phoneservice', 'paperlessbilling', 'churn']

for col in binary_cols:
    df[col] = df[col].map({'Yes': True, 'No': False})
```

```python
cols_to_clean = ['multiplelines', 'onlinesecurity', 'onlinebackup', 
                 'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies']

for col in cols_to_clean:
    df[col] = df[col].replace({'No internet service': 'No'})
```

```python
df['customerid'] = df['customerid'].astype('string')
```
