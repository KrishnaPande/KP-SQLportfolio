```python
import pandas as pd

df = pd.read_csv("Telco-Customer-Churn.csv")
df
```

```python
df.dtypes
```

```python
df['TotalCharges'].replace(" ", pd.NA, inplace=True)
```

```python
df['TotalCharges'].isna().sum()
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
for col in df.columns:
    print(f"Column: {col}")
    print(df[col].unique())
    print(len(df[col].unique()))
    print("-" * 30)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.dtypes
```

```python
yes_no_cols = ['partner', 'dependents', 'phoneservice', 'paperlessbilling', 'churn']

for col in yes_no_cols:
    df[col] = df[col].map({'Yes': True, 'No': False})
```

ðŸ’¡ Why:
Booleans are efficient and make your data easier to work with in analysis or modeling.

```python
cat_cols = ['gender', 'multiplelines', 'internetservice', 'onlinesecurity', 'onlinebackup',
            'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies',
            'contract', 'paymentmethod']

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
df['seniorcitizen'].unique()  # Should be [0, 1]
df['seniorcitizen'] = df['seniorcitizen'].astype(bool)
```

```python
df.dtypes
```

```python
df['customerid'] = df['customerid'].astype('string')
```

ðŸ§¹ Data Cleaning & Preprocessing Summary
We performed the following cleaning and preprocessing steps on the Telco Customer Churn dataset:

1. Null Values Check:

Verified there are no missing values using df.isnull().sum().

2. Data Type Fixes:

Converted customerid to string[python] (not category, as values are unique).

Converted seniorcitizen to bool (already in correct format).

Converted binary Yes/No columns to boolean:

partner, dependents, phoneservice, paperlessbilling, churn

3. Category Columns Optimization:

Converted applicable columns to category for efficient memory usage:

gender, multiplelines, internetservice, onlinesecurity, onlinebackup, deviceprotection, techsupport, streamingtv, streamingmovies, contract, paymentmethod

4. String Cleanup:

Replaced 'No internet service' and 'No phone service' with 'No' in relevant columns to simplify values.

5. Numeric Type Conversion:

Converted totalcharges to float64 for proper numerical analysis (after handling blank strings if required).

Ensured monthlycharges and tenure remain as numeric types.

6. Duplicates Check:

Confirmed no duplicate rows exist using df.duplicated().sum().
